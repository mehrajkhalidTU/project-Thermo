# Bepaling van soortelijke warmte van water


# Introductie

Water heeft een enorme buffercapaciteit: je moet veel energie toevoegen om de temperatuur van water een graad te verwarmen. In dit practicum gaan we de soortelijke warmte van water bepalen door een bekende hoeveelheid water te verwarmen met een bekende hoeveelheid energie, en de temperatuurstijging te meten.

# Theorie

basis formule voor dit experiment: De hoeveelheid warmte-energie ($Q$) die nodig is om de temperatuur van een stof te veranderen, is recht evenredig met de massa ($m$) van de stof en de temperatuurverandering ($\Delta T$). De evenredigheidsconstante is de soortelijke warmtecapaciteit ($c$).$$\text{Warmte-energie:} \quad Q = m \cdot c \cdot \Delta T$$
In dit experiment wordt de warmte-energie ($Q$) elektrisch geleverd door een verwarmingselement (weerstand) dat is aangesloten op een voedingsbron. 

De elektrische energie die wordt omgezet in warmte is gelijk aan het product van het elektrisch vermogen ($P$) en de tijd ($t$).$$\text{Elektrische energie:} \quad Q_{\text{elektrisch}} = P \cdot t$$Het elektrisch vermogen ($P$) wordt berekend uit de spanning ($V$) over de weerstand en de stroomsterkte ($I$) erdoorheen: $P = V \cdot I$.$$\text{Dus:} \quad Q_{\text{elektrisch}} = V \cdot I \cdot t$$

Door de toegevoerde elektrische energie gelijk te stellen aan de opgenomen warmte-energie van het water (idealiter, zonder warmteverlies):$$V \cdot I \cdot t = m \cdot c \cdot \Delta T$$De soortelijke warmte van water ($c$) kan dan worden berekend door de formule om te schrijven:$$c = \frac{V \cdot I \cdot t}{m \cdot \Delta T}$$

```{solution}
1. Bereken $\Delta T$: Bereken de totale temperatuurstijging $\Delta T = T_{\text{eind}} - T_{\text{start}}$.
2. Bereken $Q$: Bereken de totale toegevoegde warmte-energie met $Q = V \cdot I \cdot t$.
3. Bereken $c$: Vul alle gemeten waarden in de formule in om de soortelijke warmte van water ($c$) te bepalen:$$c = \frac{Q}{m \cdot \Delta T} = \frac{V \cdot I \cdot t}{m \cdot \Delta T}$$
```

# Resultaten


````python
# Hier de data en de analyse
import matplotlib.pyplot as plt
import numpy as np
I = 1.98  # in AmpÃ¨re
U = 19.0  # in Volt
m_maatbeker = 271.3  # in gram
m_water = 925.4 - m_maatbeker  # in gram

T_begin = 21.0  # in graden Celsius
T = np.array([21.0, 21.4, 21.8, 22.3, 22.7, 23.1, 23.5, 23.8, 24.1, 24.5, 24.8, 25.2, 25.5, 25.8, 26.1, 26.5, 26.7, 27.2, 27.5, 27.9]) #metingen met tijdsintervallen van 30 seconden
t = np.array([0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360, 390, 420, 450, 480, 510, 540, 570])  # in seconden
dT = T[1:21] - T_begin  # in graden Celsius

Q = t[1:21] * U * I
c = Q / (m_water/1000 * dT)
print(c)
# Nu de soortelijke warmte over tijd plot
plt.plot(t[1:21], c, 'o-', label='Gemeten waarde', markersize=5)
plt.plot([0, 600], [4186, 4186], 'r--', label='Theoretische waarde (4186 J/kgK)')
plt.xlabel('Tijd (s)')
plt.ylabel('Soortelijke warmte (J/kgK)')
plt.title('Soortelijke warmte van water over tijd')
plt.legend()
plt.grid()
plt.show()
# Nu de geabsorbeerde warmte over tijd plot
plt.plot(t[1:21], Q, 'r.', markersize=10)
plt.xlabel('Tijd (s)')
plt.ylabel('Geabsorbeerde warmte (J)')
plt.title('Geabsorbeerde warmte van water over tijd')
plt.grid()
plt.show()
````

````python
from scipy.optimize import curve_fit
def model(t, c):
    return (U * I * t) / (m_water/1000 * c) + T_begin
popt, pcov = curve_fit(model, t, T)
c_fit = popt[0]
c_error = np.sqrt(np.diag(pcov))[0]
print(f"Soortelijke warmtecapaciteit van water (fit): {c_fit:.3f} Â± {c_error:.3f} J/kgK")

t_test = np.linspace(0, 570, 100)
T_fit = model(t_test, c_fit)
plt.plot(t, T, 'o', label='Data punten')
plt.plot(t_test, T_fit, '-', label='Curve fit')
plt.xlabel('Tijd (s)')
plt.ylabel('Temperatuur (Â°C)')
plt.title('Temperatuurstijging van water bij verwarming')
plt.legend()
plt.grid()
plt.show()
````

wegens warmte verlies is dT kleiner dan verwacht, hierdoor zal de berekende c_fit kleiner zijn als we alle data punten nemen. de waarde komt meer overeen met de literatuurwaarde als we de tijdsinterval verkorten naar 210 seconden. 

````python
t2 = t[0:8]
T2 = T[0:8]
popt2, pcov2 = curve_fit(model, t2, T2)
c_fit2 = popt2[0]
c_error2 = np.sqrt(np.diag(pcov2))[0]
print(f"Soortelijke warmtecapaciteit van water (eerste 210 sec): {c_fit2:.3f} Â± {c_error2:.3f} J/kgK")

t2_test = np.linspace(0, 210, 100)
T2_fit = model(t2_test, c_fit2)
plt.plot(t2, T2, 'o', label='Data punten (eerste 210 s)')
plt.plot(t2_test, T2_fit, '-', label='Curve fit (eerste 210 s)')
plt.xlabel('Tijd (s)')
plt.ylabel('Temperatuur (Â°C)')
plt.title('Temperatuurstijging van water bij verwarming (eerste 210 s)')
plt.legend()
plt.grid()
plt.show()
````

````python
#strijdigheidsanalyse met literatuurwaarde
c_literatuur = 4186  # J/kgK
afwijking = abs(c_fit2 - c_literatuur)
gecombineerde_onzekerheid = c_error2
t_waarde = afwijking / gecombineerde_onzekerheid
if t_waarde <= 1:
    print(f"Volledige overeenstemming (t = {t_waarde:.2f} <= 1)")
    print("Het verschil is kleiner dan Ã©Ã©n gecombineerde standaardfout.")
elif t_waarde <= 2:
    print(f"Redelijke overeenstemming (1 < t = {t_waarde:.2f} <= 2)")
    print("De waarden zijn niet in conflict, maar het verschil is merkbaar.")
else: # t_waarde > 2
    print(f"Strijdigheid (t = {t_waarde:.2f} > 2)")
    print("De waarden zijn in conflict (niet in overeenstemming).")
print("Dus de soortelijke warmte van water gemeten binnen de tijdsinterval 210 seconden komt overeen met de literatuurwaarde.")
````

# Discussie en conclusie

De metingen geven een expirementele waarde van c = 4179 +/- 42 J/kgK, voor de soortelijke warmte van water. deze waarde is verkregen door een curve fit uit te voeren op de gekregen data bij een tijdsinterval van 210 seconden. dit is gedaan om de warmteverlies invloed te minimaliseren. uit de strijdigheidsanalyse blijkt dat de experimentele waarde overen komt met de literatuurwaarde 4186 J/kgK (met een t-waarde van 0.16). 

Als de volledige tijdsinterval tot 570 seconden werd gebruikt werd een exprimentele waarde van 4666 +/- 40 J/KgK gevonden. dit is te verklaren doordat de energie niet alleen wordt ontvangen door de water in de maatbeker en dat er warmteverlies aanwezig is van de water in de maatbeker zelf. om de invloed van deze twee factoren te minimaliseren in de analyse, is alleen rekening gehouden met de eerste 210 seconden van het experiment.
