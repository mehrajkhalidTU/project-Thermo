```{exercise}
Welke vormen van warmtetransport bestaan er?
```
```{solution}
Er zijn drie vormen van warmtetransport:

1. Geleiding (Conductie):
 Dit is warmteoverdracht door direct contact tussen deeltjes, zonder verplaatsing van het materiaal zelf (meestal in vaste stoffen).

2. Stroming (Convectie):
 Dit is warmteoverdracht door de verplaatsing van een vloeistof of gas (bijvoorbeeld warme lucht dor opstijgt).

3. Straling (Radiatie):
 Dit is warmteoverdracht via elektomagnetische golven, waarbij geen materie nodig is om de warmte te verplaatsen. 

```
In deze proef proberen we een inschatting te maken van de ordegrootte van de verschillende vormen van warmtetransport bij de koeling van een metalen buis aan lucht. 

# Theorie

Volgens [Newton's wet van afkoeling](https://nl.wikipedia.org/wiki/Afkoelingswet_van_Newton) is de snelheid waarmee een voorwerp afkoelt evenredig met het verschil in de temperatuur van het voorwerp ($T$) en de omgeving ($T_0$). We kunnen dit schrijven als:

$$
    \dot{Q} = -hA(T(t) - T_0),
$$

waarin 

- $\dot{Q}$ de warmtestroom in $\mathrm{W}$,
- $A$ het oppervlak waardoor koeling optreedt in $\mathrm{m}^2$,
- $h$ de warmteoverdrachtscoëfficiënt in $\mathrm{W/(m^2 K)}$.

Dit levert de differentiaalvergelijking

$$
    C\dot{T} = -hA(T(t) - T_0),
$$

met $C$ de warmtecapaciteit in $\mathrm{J/kg}$. Herschrijven met $\tau = \frac{C}{hA}$ levert:

$$
    -\tau\dot{T} = T(t) - T_0,
$$ (eq:dv1)

met als oplossing:

$$
    T(t) - T_0 = (T(0) - T_0)\text{e}^{-t/\tau}.
$$ (eq:dv1_opl)

We kunnen hieruit dus concluderen dat $\tau$ de karakteristieke tijdsduur is waarin de temperatuur van de buis een factor $\text{e}$ verlaagd ten opzichte van de omgevingstemperatuur. 

```{exercise} Oplossing DV controleren 
:label: ex-dvopl
Controleer bovenstaande door vergelijking [](#eq:dv1_opl) te substitueren in [](#eq:dv1).
```

```{solution} ex-dvopl

Om de bovenstaande vergelijking te controleren, worden er vier stappen doorlopen: 

Stap 1: Bepaal $T(t)$ uit vergelijking (4): 
 $$T(t) = T_0 + (T(0) - T_0)e^{-t/\tau}$$ 

Stap 2: Bepaal de tijdsafgeleide $\dot{T}$: 
 De afgeleide van $T_0$ (een constante) is 0. De afgeleide van $e^{ax}$ is $ae^{ax}$. 
 $$\dot{T} = \frac{dT(t)}{dt} = (T(0) - T_0) \cdot \left(-\frac{1}{\tau}\right) e^{-t/\tau}$$ $$\dot{T} = -\frac{1}{\tau} (T(0) - T_0)e^{-t/\tau}$$

Stap 3: Substitueer $\dot{T}$ in de linkerkant (L.K.) van vergelijking (3):
 $$\text{L.K.} = -\tau\dot{T} = -\tau \cdot \left[ -\frac{1}{\tau} (T(0) - T_0)e^{-t/\tau} \right]$$ $$\text{L.K.} = (T(0) - T_0)e^{-t/\tau}$$

Stap 4: Bepaal de rechterkant (R.K.) van vergelijking (3):
 $$\text{R.K.} = T(t) - T_0$$Volgens vergelijking (4) is:$$\text{R.K.} = (T(0) - T_0)e^{-t/\tau}$$
Aangezien $\text{L.K.} = \text{R.K.}$, is de voorgestelde oplossing (4) de correcte oplossing voor de differentiaalvergelijking (3)

```

We zijn hier voor het gemak uitgegaan van een $h$ die onafhankelijk is van de temperatuur. We weten echter dat warmtetransport door straling niet lineair gaat, maar als 

$$
    \dot{Q}_s = \epsilon \sigma A (T^4 - T_0^4).
$$ 

Voor kleine temperatuurverschillen ($\Delta T = T - T_0$) is dit te vereenvoudigen tot

$$
    \dot{Q}_s = \epsilon \sigma A ((T_0+\Delta T)^4 - T_0^4) \approx \epsilon \sigma 4A T_0^3 \Delta T.
$$

Zolang $\Delta T$ dus relatief klein is ten opzichte van $T_0$, kunnen we $h$ dus inderdaad als een constante beschouwen. 

```{exercise} Uitwerken
:label: ex_uitw
```
Laat zien dat bovenstaande geldt door eerst $(T_0+\Delta T)^4$ uit te schrijven, te substitueren en dan te bedenken dat $\Delta T$ klein is en $\Delta T^2$ dus nog kleiner.


```{solution} ex_uitw

1. Substitueer $T = T_0 + \Delta T$ in de warmtestroomvergelijking: $$\dot{Q}_s = \epsilon \sigma A \left( (T_0 + \Delta T)^4 - T_0^4 \right)$$ 
2. Schrijf de term $(T_0 + \Delta T)^4$ uit met behulp van het binomium van Newton: $$(T_0 + \Delta T)^4 = T_0^4 + 4T_0^3(\Delta T) + 6T_0^2(\Delta T)^2 + 4T_0(\Delta T)^3 + (\Delta T)^4$$ 
3. Substitueer dit terug in de formule voor $\dot{Q}_s$: $$\dot{Q}_s = \epsilon \sigma A \left[ \left( T_0^4 + 4T_0^3\Delta T + 6T_0^2(\Delta T)^2 + 4T_0(\Delta T)^3 + (\Delta T)^4 \right) - T_0^4 \right]$$ $$\dot{Q}_s = \epsilon \sigma A \left[ 4T_0^3\Delta T + 6T_0^2(\Delta T)^2 + 4T_0(\Delta T)^3 + (\Delta T)^4 \right]$$
4. Maak de aanname voor kleine $\Delta T$: Als $\Delta T$ klein is ten opzichte van $T_0$, geldt dat termen met $\Delta T^2$, $\Delta T^3$, enzovoort, verwaarloosbaar klein zijn ten opzichte van de term met $\Delta T$.$$\text{Als } |\Delta T| \ll T_0, \text{ dan geldt: } 6T_0^2(\Delta T)^2 + 4T_0(\Delta T)^3 + (\Delta T)^4 \approx 0$$
5. Vereenvoudiging: $$\dot{Q}_s \approx \epsilon \sigma 4A T_0^3 \Delta T$$
```

```{solution} ex_conf

Ja, de oriëntatie maakt uit. 

 1. Verticale oriëntatie: De lucht langs de buis warmt op, stijgt op, en creëert een stabiele, laminaire grenslaag langs de lengte van de buis. Dit bevordert efficiënte natuurlijke convectie (luchtstroom langs het oppervlak).

 2. Horizontale oriëntatie: De opgestegen warme lucht heeft de neiging om direct naar boven te bewegen, waardoor de lucht rond de onderkant van de buis heter blijft dan de lucht rond de bovenkant. Dit kan leiden tot een minder uniforme, en in sommige gevallen, minder efficiënte convectie. Over het algemeen is de warmteoverdrachtscoëfficiënt $h$ iets hoger in de verticale oriëntatie door de meer gestroomlijnde convectie. 

De dop dient om warmtetransport via de binnenkant van de buis te elimineren:
 1. Zonder dop: De buis staat in open verbinding met de lucht. Dit veroorzaakt interne convectie (lucht beweegt in en uit de buis) en straling vanaf de binnenwand. Deze interne processen zijn moeilijk te modelleren en zouden de gemeten warmteoverdrachtscoëfficiënt $h$ vervalsen.
 
 2. Met dop: De lucht binnenin de buis wordt opgesloten en koelt mee af met de buis. Warmtetransport vindt dan alleen plaats via het uitwendige oppervlak van de metalen buis, wat overeenkomt met de aannames in de theorie (vergelijking 1: $\dot{Q} = -hA(T(t) - T_0)$).
```

# Warmteoverdrachtsmechanismen
 
 De warmteoverdracht wordt gedomineerd door convectie. De lucht rond de buis warmt op en stijgt op. Straling speelt een kleinere rol bij deze relatief lage temperaturen. Geleiding wordt verwaarloosd, aangezien de buis nauwelijks contact maakt met andere vaste stoffen.
 
     Mogelijke foutbronnen
     
         Verdamping: Als de buis aan de bovenkant open was, is er warmte verloren gegaan door verdamping (latente warmte). Dit zorgt voor een snellere afkoeling en dus een overschatting van $h$.
         
         Temperatuurverdeling: Er is aangenomen dat de temperatuur in het water overal gelijk is ($Bi < 0,1$). Als er niet geroerd is, was de kern mogelijk warmer dan de wand, wat de meting beïnvloedt.

# Discussie en conclusie


Het doel van dit experiment was het bepalen van de karakteristieke afkoeltijd en de warmteoverdrachtscoëfficiënt van een cilinder met water. 

Op basis van de metingen en de exponentiële fit is vastgesteld dat: 

 1-De karakteristieke tijd ($\tau$) ongeveer 4597 seconden bedraagt.

 2-De warmteoverdrachtscoëfficiënt ($h$) ongeveer 9,1 W/(m²·K) bedraagt.
 
Hoewel de wiskundige fit door de korte meetduur een onrealistische asymptoot vertoonde, komt de gevonden $h$-waarde goed overeen met de theorie voor natuurlijke convectie in lucht. Hiermee is aangetoond dat de afkoeling in de eerste fase consistent is met de afkoelingswet van Newton.
